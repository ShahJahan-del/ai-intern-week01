```python
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)

# Show first 5 rows (test)
df.head()
```

```python
### Exploratory Data Analysis ###

# Show number of rows, columns and data types (ex: float, object...)
df.info()

# Automatic statistics summary (Mean, Median/50%, std dev, etc.)
df.describe()

# Number of missing values (NaN) per column
df.isnull().sum()
```

```python
### Data Preprocessing and Feature Engineering ###

## 1. Data cleaning

## Column: Age (Numerical)
# Strategy : Replace by median, because there are outliers in the distribution
age_median = df['age'].median()
df['age'] = df['age'].fillna(age_median)

## Column: Embark Town (Categorical)
# Strategy : Replace by the mode (most frequent value), because there are only 2 NAs in this column
town_mode = df['embark_town'].mode()[0]
df['embark_town'] = df['embark_town'].fillna(town_mode)

## Column: 'embarked' (shortened version of the embark town)
embarked_mode = df['embarked'].mode()[0]
df['embarked'] = df['embarked'].fillna(embarked_mode)

## Column: Deck (Categorical)
# Strategy : Create a 'Unknown' value, because there are too many NAs in this column (688/891)
df['deck'] = df['deck'].fillna('Unknown')

# Checking if there are any remaining NAs after cleaning
print("Summary of NAs after cleaning")
print(df[['age', 'embark_town', 'embarked', 'deck']].isnull().sum())
print("-" * 35)




## 2. Encoding categoricals

## Sex (Binary -> Mapping 0/1)
# Manually mapping sex category
df['sex_encoded'] = df['sex'].map({'male': 1, 'female': 0})

## Every other categorical columns
# Strategy : Automatic One-Hot Encoding (create a category for every textual value and create binary
# columns for them)
# Example : The value 'Cherbourg' for the column 'embark_town' becomes a column 'embark_town_Cherbourg'
# which can have binary values 0 (no) or 1 (yes)
columns_to_encode = ['embarked', 'class', 'who', 'deck', 'embark_town', 'alive']

# pd.get_dummies automatically creates binary columns (0 or 1)
df = pd.get_dummies(df, columns=columns_to_encode, dtype=int)

# --- Aperçu du résultat final ---
print("\nVariables available for Machine Learning :")
print(df.dtypes)
df.head()
```

```python
### Encoding categoricals ###

## Sex

# Replace text by integers (mapping)
df['sex_encoded'] = df['sex'].map({'male': 1, 'female': 0})

# Show a preview (test)
df[['sex', 'sex_encoded']].head()
```

```python
import matplotlib.pyplot as plt

# 1. Age histogram
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of age for the passengers of the Titanic')
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.grid(axis='y', alpha=0.75)
plt.savefig('histogram_age.png')  # Save file as PNG
plt.show()

# 2. Scatter plot (cloud of points) : Fare vs Age
plt.figure(figsize=(8, 5))
plt.scatter(df['age'], df['fare'], alpha=0.5, color='darkorange', edgecolor='none')
plt.title('Fare related to age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True, alpha=0.3)
plt.savefig('scatter_fare_vs_age.png')  # Save file as PNG
plt.show()
```
